```python
# import
import seaborn as sns
import pandas as pd
sns.set_theme()
tweets = pd.read_csv("covid-19_vaccine_tweets_with_sentiment.csv")
vax = pd.read_csv("vaccination_all_tweets.csv")
print(tweets.shape)
tweets.head()
vax.head()
```

```python
# graphs
# merge
merged = pd.merge(tweets, vax, left_on = 'tweet_id', right_on = 'id')
merged.head()
```

```python
bar = sns.catplot(x = 'label', y = 'user_followers', kind = 'bar', data = merged)
bar
box = sns.catplot(x = 'label', y = 'user_followers', kind = 'box', data = merged)
box
```

```python
bar = sns.catplot(x = 'label', y = 'retweets', kind = 'bar', data = merged)
bar
box = sns.catplot(x = 'label', y = 'retweets', kind = 'box', data = merged)
box
```

```python
line = sns.relplot(x = 'date', y = 'retweets', data = merged, hue = 'label', kind = 'line')
line
merged.head()
```

```python
import numpy as np
early = merged.iloc[np.where(merged['date'] < '2021-01-18 ')]
early
```

```python
import numpy as np
many_retweets = merged.iloc[np.where(merged['retweets'] > 100)]
many_retweets
```

```python
many_rt_line = sns.relplot(x = 'date', y = 'retweets', data = many_retweets, hue = 'label', kind = 'line')
many_rt_line
```

```python
copy = many_retweets.copy()
many_retweets.index = pd.to_datetime(many_retweets['date'],format='%Y-%m-%d %H:%M:%S')
print(many_retweets)
x = many_retweets.groupby(by=[many_retweets.index.month, many_retweets.index.year]).count()
x
```

```python
merged.index = pd.to_datetime(merged['date'],format='%Y-%m-%d %H:%M:%S')
y = merged.groupby(by=[merged.index.month, merged.index.year]).count()
y
```

```python
ones = merged.iloc[np.where(merged['label'] == 1)]
grouped_ones = ones.groupby(by=[ones.index.month, ones.index.year]).count()
grouped_ones.head()
```

```python
bar_ones = sns.catplot(x = y.index.values, y = 'retweets', kind = 'bar', data = grouped_ones)
bar_ones
```

```python
twos = merged.iloc[np.where(merged['label'] == 2)]
grouped_twos = twos.groupby(by=[twos.index.month, twos.index.year]).count()
threes = merged.iloc[np.where(merged['label'] == 3)]
grouped_threes = threes.groupby(by=[threes.index.month, threes.index.year]).count()
```

```python
bar_twos = sns.catplot(x = y.index.values, y = 'retweets', kind = 'bar', data = grouped_twos)
bar_twos
bar_threes = sns.catplot(x = y.index.values, y = 'retweets', kind = 'bar', data = grouped_threes)
bar_threes
```

```python
grouped_ones['1'] = grouped_ones['label']
grouped_ones['hue'] = '1'
grouped_twos['2'] = grouped_twos['label']
grouped_twos['hue'] = '2'
grouped_threes['3'] = grouped_threes['label']
grouped_threes['hue'] = '3'
```

```python
all_sentiments = pd.concat([grouped_ones, grouped_twos, grouped_threes])
all_sentiments_grouped = all_sentiments.groupby([all_sentiments.index.values, 'hue']).sum()
all_sentiments_grouped.reindex(['(12, 2020)', '(1, 2021)', '(2, 2021)', '(3, 2021)', '(4, 2021)'])
print(all_sentiments_grouped)
```

```python
newest = sns.catplot(x = all_sentiments.index.values, y = 'retweets', kind = 'bar', data = all_sentiments, hue = 'hue', col_order = ['(12, 2020)', '(1, 2021)', '(2, 2021)', '(3, 2021)', '(4, 2021)'])
newest.set(xlabel='date')
newest
```

```python
print(all_sentiments_grouped)
grouped_new = sns.catplot(x = all_sentiments_grouped.index.values, y = 'retweets', kind = 'bar', data = all_sentiments_grouped)
grouped_new.set(xlabel='date')
grouped_new
```

```python
months_line = sns.relplot(x = all_sentiments_grouped.index.values, y = 'retweets', data = all_sentiments_grouped, hue = 'hue', kind = 'line')
months_line
```
