```python
# Run this code cell to import relevant libraries
# Please update to seaborn v0.11 for this assignment
# check version: print(sns.__version__)

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set_theme()
```

```python
# Run but do not modify this code
# Only 11 penguins out of over 300 are missing any data;
# we simply drop those here
# Run but do not modify this code
vaccinations = pd.read_csv("owid-covid-data.csv")
vaccinations['date'] = pd.to_datetime(vaccinations['date'], format='%Y-%m-%d')
regions = pd.read_csv("all.csv")
vaccinations = pd.merge(vaccinations, regions, left_on='iso_code', right_on='alpha-3')
vaccinations.head()
print(set(vaccinations['sub-region'].values))
```

```python
#Eastern Asia
vacMap = vaccinations.loc[vaccinations['sub-region'] == 'Eastern Asia']
saveStr = 'Eastern Asia' + ".png"
sns.set_theme(rc = {'figure.figsize':(8,8)})
sns.lineplot(data=vacMap, x="date", y="total_vaccinations_per_hundred", hue = "iso_code", 
             legend = 'brief').set(title = 'Eastern Asia')
plt.savefig(saveStr, dpi=300)
```
